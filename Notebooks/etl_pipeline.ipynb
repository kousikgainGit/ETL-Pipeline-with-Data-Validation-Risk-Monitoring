# ETL Pipeline with Data Validation & Risk Monitoring

```
import pandas as pd
import numpy as np
```

## Extract: Load Raw Data

```
df = pd.read_csv('data/raw_sales_data.csv')
df.head()
```

## Initial Data Overview

```
df.info()
df.describe(include='all')
```

## Transform: Data Cleaning

```
df['quantity'] = df['quantity'].fillna(df['quantity'].median())
df['price'] = df['price'].fillna(df['price'].median())
df['payment_method'] = df['payment_method'].fillna('Unknown')
df['transaction_date'] = pd.to_datetime(df['transaction_date'])
```

```
df = df.drop_duplicates()
```

## Feature Engineering

```
df['revenue'] = df['quantity'] * df['price']
```

## Data Validation Checks

```
missing_values = df.isnull().sum()
duplicate_count = df.duplicated().sum()
invalid_quantity = df[df['quantity'] <= 0]
invalid_price = df[df['price'] <= 0]

missing_values, duplicate_count, len(invalid_quantity), len(invalid_price)
```

## Risk Flagging

```
def risk_flag(row):
    if row['payment_method'] == 'Unknown':
        return 'Missing Payment Method'
    elif row['revenue'] > 50000:
        return 'High Value Transaction'
    else:
        return 'Normal'

df['risk_flag'] = df.apply(risk_flag, axis=1)
df[['transaction_id','revenue','risk_flag']].head()
```

## Load: Save Cleaned Data

```
df.to_csv('data/cleaned_data.csv', index=False)
```

## Summary Insights

```
print("Total Transactions:", len(df))
print("Total Revenue:", df['revenue'].sum())
print("Risky Transactions:", (df['risk_flag'] != 'Normal').sum())
```
